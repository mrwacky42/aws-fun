```python
from pprint import pprint as pp
import numpy as np
import pandas as pd
import scipy
from collections import Counter
import matplotlib.pyplot
from IPython.display import SVG
import pygal
import gzip
%matplotlib inline
```

```python
import reserved_instances as ri
```

```python
pricing = dict()
for option in ri.redshift_pricing.keys():
    pricing[option] = ri.aggregate_costs(ri.generate_monthly_costs(option, 36))
```

```python
line_chart = pygal.Line()
line_chart.title = 'Cash flow analysis'
line_chart.x_labels = map(str, range(0, 36))
for option in pricing.keys():
    prices = [x*34 for x in pricing[option]]
    line_chart.add(option, prices)
SVG(line_chart.render())
```
